```python
def oddOrEven(x):
   if x%2 == 1:
       print("it is an odd number")
   else:
       print("it is an even number")
        
```

```python
def OddEven(x):
    if isinstance(x, int):
        if x%2 == 0:
            if x == 0:
                statement = " is neither even or odd"
            else:
                statement = " is even number"
        else:
            statement = " is odd number"
    else:
        statement = " is not an integer"
        
    return print(f"{x}" + statement)
```

```python
oddOrEven(12)
```

```python
OddEven(0)
```

```python
OddEven(9.7)
```

```python
OddEven(4)
```

```python
# importing system module
import sys
  
# declaring a list of 1000 elements
S= range(1000)
  
# printing size of each element of the list
print("Size of each element of list in bytes: ",sys.getsizeof(S))
  
# printing size of the whole list
print("Size of the whole list in bytes: ",sys.getsizeof(S)*len(S))
  
# declaring a Numpy array of 1000 elements
D= np.arange(1000)
  
# printing size of each element of the Numpy array
print("Size of each element of the Numpy array in bytes: ",D.itemsize)
  
# printing size of the whole Numpy array
print("Size of the whole Numpy array in bytes: ",D.size*D.itemsize)
```

```python
#Q3: What do you notice? Propose a way to reassign arr_slice with new value without modifying my_arr.
my_arr = np.arange(10)
print(my_arr)
my_arr[4:7] = 25
print(my_arr)
arr_slice = my_arr[4:7]
# Change the first element of arr_slice to -1
arr_slice[0]= -1
print(arr_slice)
print(my_arr)
print()

# modify the element's value of arr_slice that copy from my_arr modified 
# the original value of my_arr accordingly as well.
my_arr1 = np.arange(10)
print("my_arr1: ", my_arr1)
my_arr1[4:7] = 25
print("\nmy_arr1 after changed the value to 25: ", my_arr1)
arr_slice1 = np.copy(my_arr1[4:7])
print("arr_slice1 before editing: ", arr_slice1)

# Change the first element of arr_slice1 to -1
arr_slice1[0]= -1
print("arr_slice1 after editing: ", arr_slice1)
print("my_arr1 after editing arr_slice1: ", my_arr1)

# Change the value of element in arr_slice1 do not change the original value of my_arr1 array
# because arr_slice1 is a copy array of my_arr1, they are two different 
# numpy arrays who are not related by reference.
```

```python
#Q4: Create an image as shown as the following with the help of Numpy and matplotlib modules. 
#    You canarbitrarily set the dimension of the image and white circular spot at the middle
import cv2
blackbackground = np.zeros((200, 200), dtype = np.uint8)
#background[50: -50, 50: -50] = 1 #white circle

cv2.circle(blackbackground, center= (100,100), radius = 50, color = (255,255,0), thickness = -1)


plt.imshow(blackbackground, cmap=plt.cm.gray)
plt.title('custom white image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mask = np.zeros((10,10))
mask[3:-3, 3:-3] = 1 # white square in black background
im = mask + np.random.randn(10,10) * 0.01 # random image
masked = np.ma.masked_where(mask == 0, mask)

import matplotlib.pyplot as plt
plt.figure()
plt.subplot(1,2,1)
plt.imshow(im, 'gray', interpolation='none')
plt.subplot(1,2,2)
plt.imshow(im, 'gray', interpolation='none')
plt.imshow(masked, 'jet', interpolation='none', alpha=0.7)
plt.show()
```
